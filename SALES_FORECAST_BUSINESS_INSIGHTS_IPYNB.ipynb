```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression

print("All libraries loaded successfully!")
```

```python
import pandas as pd

sales_data = {
    "Month": [
        "Jan", "Feb", "Mar", "Apr", "May", "Jun",
        "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"
    ],
    "Sales": [
        12000, 13500, 14200, 15000, 16500, 17500,
        18200, 19000, 20500, 22000, 23500, 25000
    ]
}

df = pd.DataFrame(sales_data)

print("Historical Sales Data")
display(df)
```

```python
plt.figure(figsize=(10,5))

plt.plot(df["Month"], df["Sales"], marker="o")

plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.grid(True)

plt.show()
```

```python
# Create month numbers
df["Month_Number"] = range(1, 13)

# Input (X) and Output (y)
X = df[["Month_Number"]]
y = df["Sales"]

# Train the model
model = LinearRegression()
model.fit(X, y)

print("AI Sales Forecasting Model Trained Successfully!")
```

```python
# Future months to predict
future_months = pd.DataFrame({
    "Month_Number": [13, 14, 15, 16]
})

# Generate predictions
predictions = model.predict(future_months)

# Create forecast table
forecast_df = pd.DataFrame({
    "Future_Month": ["Month 13", "Month 14", "Month 15", "Month 16"],
    "Predicted_Sales": predictions.astype(int)
})

print("Future Sales Forecast")
display(forecast_df)
```

```python
plt.figure(figsize=(12,6))

# Historical Sales
plt.plot(
    df["Month_Number"],
    df["Sales"],
    marker="o",
    label="Historical Sales"
)

# Forecasted Sales
plt.plot(
    future_months["Month_Number"],
    predictions,
    marker="o",
    linestyle="--",
    label="Forecasted Sales"
)

plt.title("Sales Forecast & Business Insights Dashboard")
plt.xlabel("Month Number")
plt.ylabel("Sales")
plt.legend()
plt.grid(True)

plt.show()
```

```python
current_sales = df["Sales"].iloc[-1]
predicted_sales = int(predictions[-1])

growth_percent = ((predicted_sales - current_sales) / current_sales) * 100

print("BUSINESS INSIGHTS")
print("-" * 40)

print(f"Latest Recorded Sales: {current_sales}")
print(f"Forecasted Sales (Month 16): {predicted_sales}")
print(f"Expected Growth: {growth_percent:.2f}%")

if growth_percent > 10:
    print("\nInsight: Strong sales growth is expected.")
    print("Recommendation: Consider increasing inventory and marketing efforts.")
else:
    print("\nInsight: Sales growth is moderate.")
    print("Recommendation: Monitor performance and optimize sales strategy.")
```

```python
plt.figure(figsize=(12,6))

# Historical Sales
plt.plot(
    df["Month_Number"],
    df["Sales"],
    marker="o",
    label="Historical Sales"
)

# Forecasted Sales
plt.plot(
    future_months["Month_Number"],
    predictions,
    marker="o",
    linestyle="--",
    label="Forecasted Sales"
)

plt.title("Sales Forecast & Business Insights Dashboard")
plt.xlabel("Month Number")
plt.ylabel("Sales")
plt.legend()
plt.grid(True)

plt.savefig("sales_forecast_dashboard.png", bbox_inches="tight")

plt.show()

print("Dashboard image saved successfully!")
```
